```python
# replacing 'Not Available', 'unknown', these values as nan.
df['Ship Mode'] = df['Ship Mode'].replace(['Not Available', 'unknown'], np.nan)
```

```python
df['Ship Mode'].unique()
```

```python
# renaming columns to follow naming convention
df.columns=df.columns.str.replace(' ','_')
df.head()
```

```python
# create a new column using 'list_price','discount_percent'-Discount
df['Discount'] = df['List_Price'] * df['Discount_Percent'] * .01
```

```python
#  Now Create Sale Price column
df['Sale_Price']=df['List_Price'] - df['Discount']
df.head()
```

```python
# Create profit column
df['Profit']=  df['Sale_Price'] - df['cost_price']
```

```python
df.drop(columns=['cost_price','List_Price','Discount_Percent'],inplace=True)
```

```python
# Changing "Order_Date" Column Datatype
df['Order_Date']=pd.to_datetime(df['Order_Date'],format='%Y-%m-%d')
```
